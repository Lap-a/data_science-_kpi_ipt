### Список

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.tail = None

    def empty(self):
        return self.tail is None

    def append(self, key):# at the end
        new_node = Node(key)
        if self.empty():
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    def print_list(self):
        if self.empty():
            print ("no elements")
        else:
            current = self.tail
            text = ""
            while current:
                text += str(current.key) + " "
                current = current.prev
            return text

    def remove(self): #obj not deleted. u just cannot see it
        if self.empty():
            print ("no elements")
        elif self.tail.prev == None:
            self.tail = None
        else:
            self.tail.prev.next = None
            self.tail = self.tail.prev

    def find_by_index(self, index):
        current = self.head
        count = 0
        while current:
            if count == index:
                return current.key
            count += 1
            current = current.next
        return "out_of_range"


llist = DoublyLinkedList()
```

```python
llist.append(8)
llist.print_list()
```

```python
llist.remove()
llist.print_list()
```
